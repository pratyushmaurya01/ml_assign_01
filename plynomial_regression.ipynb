# Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducibility
np.random.seed(42)

# Generate X
X = np.linspace(-3, 3, 30).reshape(-1, 1)

# True function + noise
y = X**2 + np.random.normal(0, 1.5, size=X.shape)

# Plot raw data
plt.scatter(X, y)
plt.title("Original Data: y = x^2 + noise")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

## Using Neural Network 

```python
# Deep neural network (intentionally complex)
model = Sequential([
    Dense(128, activation='relu', input_shape=(1,)),
    Dense(128, activation='relu'),
    Dense(128, activation='relu'),
    Dense(1)
])

model.compile(
    optimizer=Adam(learning_rate=0.01),
    loss='mse'
)

model.fit(X, y, epochs=800, verbose=0)

# Dense X for smooth curve
X_dense = np.linspace(-3, 3, 300).reshape(-1, 1)
y_pred_nn = model.predict(X_dense)
```

```python
# Plot
plt.scatter(X, y, label="Data")
plt.plot(X_dense, y_pred_nn, color='red', label="Deep NN ")
plt.title("Overfitting with Deep Neural Network")
plt.legend()
plt.show()
```

## Using Polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Linear regression on polynomial features
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# Predict
X_dense_poly = poly.transform(X_dense)
y_pred_poly = poly_model.predict(X_dense_poly)
```

```python
plt.scatter(X, y)
plt.plot(X_dense, y_pred_poly, color='green', linewidth=3)
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```
